```python
tasks = []

while True:
    print("\n1.Add  2.View  3.Update  4.Delete  5.Exit")
    ch = input("Choice: ")

    if ch == '1':
        task = input("Enter task: ")
        tasks.append({"task": task, "status": "Pending"})
        print("Task added")

    elif ch == '2':
        for i, t in enumerate(tasks, 1):
            print(i, t["task"], "-", t["status"])

    elif ch == '3':
        i = int(input("Task number: ")) - 1
        tasks[i]["status"] = "Completed"
        print("Task updated")

    elif ch == '4':
        i = int(input("Task number: ")) - 1
        tasks.pop(i)
        print("Task deleted")

    elif ch == '5':
        print("Goodbye!")
        break

    else:
        print("Invalid choice")
```
